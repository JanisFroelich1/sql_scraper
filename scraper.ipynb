```python
import pandas as pd

# URL of the Parquet file
url_test = "https://huggingface.co/datasets/gretelai/synthetic_text_to_sql/resolve/main/synthetic_text_to_sql_test.snappy.parquet"
url_train = "https://huggingface.co/datasets/gretelai/synthetic_text_to_sql/resolve/main/synthetic_text_to_sql_train.snappy.parquet"

# Read the Parquet file from the URL
df_test = pd.read_parquet(url_test, engine='fastparquet')
df_train = pd.read_parquet(url_train, engine='fastparquet')
```

```python
print(f"sql task type: {df_train.sql_task_type.unique()}")

print(f"sql complexity description: {df_train.sql_complexity_description.unique()}")

print(f"sql task type description: {df_train.sql_task_type_description.unique()}")
```

```python
# create text data for training
# Extract the necessary columns
sql_prompts = df_train['sql_prompt']
sql_answers = df_train['sql']

# Create a text file and write the sql_prompts and sql_answers
with open('sql_prompts_and_answers.txt', 'w') as file:
    for prompt, answer in zip(sql_prompts, sql_answers):
        file.write(f"Prompt: {prompt}\n")
        file.write(f"SQL: {answer}\n\n")

# Display the path to the created file
import os
file_path = os.path.abspath('sql_prompts_and_answers.txt')
```
